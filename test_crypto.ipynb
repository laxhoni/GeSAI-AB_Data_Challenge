```python
import sqlite3

# Conectamos a la base de datos "a lo bruto" (sin pasar por la app)
conn = sqlite3.connect('gesai.db')
cursor = conn.cursor()

print("--- LO QUE VE UN HACKER ---")

# 1. Mirar un cliente
row = cursor.execute("SELECT nombre, email FROM clientes LIMIT 1").fetchone()
print(f"Nombre: {row[0]}")
print(f"Email:  {row[1]}")

# 2. Mirar la contraseña del Admin
admin = cursor.execute("SELECT contrasena FROM usuarios_empresa WHERE email='empresa@gesai.com'").fetchone()
print(f"Password Admin: {admin[0]}")

conn.close()
```
